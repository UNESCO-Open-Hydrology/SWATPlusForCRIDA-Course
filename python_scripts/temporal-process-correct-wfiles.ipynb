```python
# Imports

import pandas as pd
import os

```

```python
path = r"C:\PHD_local_asus\CRIDA\SWATPlusForCRIDA\katari-swat-crida-spt\scenarios\Toolbox\Calibration"
save_path = r"C:\PHD_local_asus\CRIDA\SWATPlusForCRIDA\weather_OK"

fileList = os.listdir(path)

pcp_list = [item for item in fileList if ".pcp" in item]
tmp_list = [item for item in fileList if ".tmp" in item]
```

```python
lats = []
lons = []
names = []
elevs = []

for item in pcp_list:
    fname = f"{path}/{item}"
    name  = item[:-4]
    
    names.append(name)
    
    with open(fname,"r") as file:   # Here we are looking for the row with variable names in txt file
        for current_line_number,line in enumerate(file,start=1):
            if current_line_number==3:
                header_line=line.strip().split()
                break
            
            
    lats.append(float(header_line[2])) 
    lons.append(float(header_line[3])) 
    elevs.append(float(header_line[4]))
    
    df=pd.read_csv(fname,
        skiprows=3,
        sep="\s+",
        header=None)
    df.columns = ["year","doy","value"]
    df['time'] = pd.to_datetime(df['year'].astype(str) + df['doy'].astype(str), format='%Y%j')
    df = df[['time','value']]
    
    df.to_csv(f"{save_path}/pcp/{name}.csv",index=None)
    
    print(f"Saved {name}.csv")    
```

```python
pcp_df = pd.DataFrame({"NAME":names,"LAT":lats,"LONG":lons,"ELEVATION":elevs})
pcp_df.to_excel(f"{save_path}/pcp_stations.xlsx",index=None)
```
